```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
from google.colab import files
files.upload()
file_path = 'age_networth.csv'
df = pd.read_csv(file_path)

# Display the first few rows
print(df.head())

# Calculate correlation
correlation = df['Age'].corr(df['Net Worth'])
print(f"Correlation between Age and Net Worth: {correlation:.2f}")

# Create scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(x='Age', y='Net Worth', data=df, scatter_kws={'s': 100})
plt.title('Correlation Between Age and Net Worth')
plt.xlabel('Age')
plt.ylabel('Net Worth')
plt.grid(True)
plt.show()
```
